```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.worldometers.info/energy/"
```

```python
res=requests.get(url)
```

```python
res
```

```python
res.content
```

```python
soup=BeautifulSoup(res.content,"html.parser")
```

```python
title=soup.title
title.text
```

```python
title.text
```

```python
Energy=soup.find("table")
Energy
```

```python
E1=Energy.find("tbody")
E1
```

```python
E2=E1.find_all("tr")
E2
```

```python
info=[]
for i in E2:
  cols=i.find_all("td")
  country_info=[j.text for j in cols]
  info.append(country_info)
print(info[:6])
```

```python
num=[]
for i in info:
  num.append(i[0])
print(num[:6])

```

```python
country=[]
for i in info:
  country.append(i[1])
print(country[:6])
```

```python
Energy_Consumption=[]
for i in info:
  Energy_Consumption.append(i[2])
print(Energy_Consumption[:6])
```

```python
World_Share=[]
for i in info:
  World_Share.append(i[3])
print(World_Share[:6])
```

```python
Per_capita_Yearly_BTU = []
for i in info:
    Per_capita_Yearly_BTU.append(i[4])
print(Per_capita_Yearly_BTU[:6])
```

```python
c_data={
    "County_name":country,
    "EC":Energy_Consumption,
    "WS":World_Share,
    "BTU":Per_capita_Yearly_BTU
}
print(c_data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(c_data)
print(df)
```
